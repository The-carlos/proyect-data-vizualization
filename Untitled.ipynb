### De las líneas de arriba puedo concluir que las siguientes columnas comparten un valor nulo [-1.0, -1 y '-1'] (Rating, Founded, Type of ownership, Industry, Sector, Competitors, Easy Apply), para que podamos identificarlos como nulos debemos reemplazar estos valores por un np.nan. Vamos a hacerlo:

```python
data = data.replace(-1, np.nan)
data = data.replace(-1.0, np.nan)
data = data.replace('-1', np.nan)

#Verificamos que esto se haya realizado con éxito:
data.isnull().sum()
```

```python
#Las únicas variables que podrían preocupar por el hecho de que son nulas son "Competitors" e "Easy Apply"
#Para el caso de la columna "Easy Apply" podemos pensar que si es negativa debe ser porque la empresa no está contratando
# De cualquierforma es más práctico manejar esta columna con un boolean:

data['Easy Apply']=data['Easy Apply'].fillna(False).astype('bool')
```

## Posiciones mejor pagadas.

```python
data
```

```python
#Primero obtenemos los salarios promedios haciendo DataCleaning

salary = list(dict(data['Salary Estimate']).values())

minim = [' '.join(re.findall('\d+K-',str(element))) for element in salary]

maxim = [' '.join(re.findall('\W\W+\d+K',str(element))) for element in salary]
#print(minim)

minim_kclean = [re.sub('K-', '', element) for element in minim]
maxim_kclean = [re.sub('K', '', element) for element in maxim]
maxim_others = [re.sub('\W\W', '', element) for element in maxim_kclean]
#print(minim_kclean)

minim_kclean = [re.sub('', '0', element) for element in minim_kclean]
maxim_others = [re.sub('', '0', element) for element in maxim_others]
#print(maxim_others)

minim = [int(element) for element in minim_kclean]
maxim = [int(element) for element in maxim_others]
#print(len(minim))

avr_salary = [((int(minim[element]) + int(maxim[element])/2)) for element in range(len(minim))]
#print(len(avr_salary))
data.drop('Salary Estimate', axis = 1, inplace = True)
data['Salary Estimate'] = avr_salary
print(data['Salary Estimate'])
```

```python
#Despues agrupamos por Job Title y obtenemos el salario promedio para las posiciones


jobs_payment = data.groupby(['Job Title']).agg('mean')['Salary Estimate']
jobs_payment = jobs_payment.sort_values(ascending=False)
jobs_payment
```

```python
#Exportamos el DataSet
jobs_payment.to_excel('salarios a aspirar.xlsx')
```

```python
#payment = pd.DataFrame(jobs_payment, columns = ['Job Title', 'Salary Estimate'])
payment = pd.DataFrame(jobs_payment.reset_index())

payment = payment.iloc[:40,]

print(payment)
#tamaños
plt.figure(figsize=(20,10))
#Agregar titulo
plt.title('Posiciones mejor pagadas.', fontsize = 14)

plt.xticks(rotation = 'vertical')
plt.bar(payment['Job Title'], payment['Salary Estimate'])
plt.ylabel('Salarios en miles de dolares al año.')
plt.show()
```
